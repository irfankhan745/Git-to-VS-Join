## Calculate Batsman Runs and Balls Faced

### Subtask:
Calculate the total runs scored by each batsman and the total number of legal balls faced by each batsman. Legal balls are those that are not wide (isWide) or no-balls (isNoBall), as these are not considered balls faced by the batsman.


**Reasoning**:
The subtask requires calculating total runs and legal balls faced by each batsman. I will filter the `ipl` DataFrame for legal deliveries (where `isWide` and `isNoBall` are not 1), then group by 'batsman' to sum `batsman_runs` for total runs and count the filtered rows for total balls faced.



```python
legal_deliveries = ipl[(ipl['isWide'].fillna(0) == 0) & (ipl['isNoBall'].fillna(0) == 0)]

total_runs = legal_deliveries.groupby('batsman')['batsman_runs'].sum()
balls_faced = legal_deliveries.groupby('batsman').size()

print("Total Runs Scored by each Batsman:\n", total_runs.head())
print("\nTotal Legal Balls Faced by each Batsman:\n", balls_faced.head())
```

# Task
The user wants to analyze the percentage of runs scored from boundaries (fours and sixes) for batsmen in the IPL.

To achieve this, I will perform the following steps:
1.  **Calculate Total Runs and Balls Faced for each Batsman**: I'll filter out wide and no-balls to count legal deliveries and sum up `batsman_runs` for each batsman.
2.  **Calculate Boundary Runs**: I'll sum `batsman_runs` only when `batsman_runs` is 4 or 6, grouping by `batsman`.
3.  **Merge Data and Filter**: I'll combine the total runs, total balls, and boundary runs into a single DataFrame. Then, I'll filter this DataFrame to include only batsmen who have faced at least 100 balls.
4.  **Calculate Boundary Percentage**: For the filtered batsmen, I'll calculate `(boundary_runs / total_runs) * 100`. I'll ensure to handle any potential division by zero.
5.  **Identify and Display Top 10 Batsmen**: I'll sort the batsmen by their `Boundary_Percentage` in descending order and display the top 10.
6.  **Visualize Boundary Percentage**: I'll create a bar chart to visually represent the `Boundary_Percentage` for these top 10 batsmen.

```python
#legeal Balls
legeal_balls = ipl[(ipl['isWide'].fillna(0)==0) & (ipl['isNoBall'].fillna(0)==0)]

#total runs after legeal balls
total_runss = legeal_balls.groupby('batsman')['batsman_runs'].sum()
#total ball faced by batsman
ball_facedd = legeal_balls.groupby('batsman').size()

#boundery runs
boundry_runs = legeal_balls[legeal_balls['batsman_runs'].isin([4,6])].groupby('batsman')['batsman_runs'].sum()


#create a dataframe for all new columns
batsman_history = pd.DataFrame({'total_runs':total_runss,'ball_faced':ball_facedd,'boundry_runs':boundry_runs})

#filterd_batsman
filtred_batsmens = batsman_history[batsman_history['ball_faced']>= 100]


#boundary precentag
filtred_batsmens = filtred_batsmens.copy()
filtred_batsmens['boundr_precentage'] = (filtred_batsmens['boundry_runs'] / filtred_batsmens['total_runs']).fillna(0) * 100


# Top 10 batsmans
top_10_bats = filtred_batsmens.sort_values(by = 'boundr_precentage',ascending=False).head(10)
print('Top 10 batsmens: ',round(top_10_bats[['boundr_precentage']]))
```
